```python
# read shapefile
import geopandas as gpd
```

```python
species_df=gpd.read_file('new_species/ll_gda2020/esrishape/whole_of_dataset/victoria/FLORAFAUNA1/HDM_FAUNA_10M.shp')
species_df.head()
```

```python
len(species_df)
```

```python
latitude=[species_df.iloc[i].geometry.centroid.y for i in range(len(species_df))]

longitude=[species_df.iloc[i].geometry.centroid.x for i in range(len(species_df)) ]
```

```python
species_df['latitude']=latitude
species_df['longitude']=longitude
```

```python
species_df.head(10)
```

```python
map_sp_df=species_df.groupby(['latitude','longitude']).size().reset_index(name='counts')
```

```python
map_sp_df.to_csv('map_sp_df.csv',index=False)
```
